# Plotting med Python

Det er mange muligheter til å lage både fine og interaktive plott med Python. Vi starter forsiktig med å plotte punkter som en polylinje.

Vi trenger biblioteket `matplotlib.pyplot` for å tegne plott, og så trenger vi noen verdier.

````python
import matplotlib.pyplot as plt
x_data = [1,2,3,4,5]
y_data = [1,4,9,16,25]

plt.plot(x_data,y_data)
````

Vi ser at det blir noen knekkpunkter ved hvert punkt. Det kan også se ut til at disse punktene er punkt på funksjonen $y=x^2$. For å klare å tegne dette som en glatt kurve, så trenger vi mange datapunkt. 

Vi kan da bruke ```numpy``` for å generere lister med mange flere verdier.

````python
import numpy as np
import matplotlib.pyplot as plt

x_verdier=np.arange(0,5,0.1) # her lages det verdier fra og med 0 til og med 5, med 0.1 som mellomrom
y_verdier=x_verdier**2
plt.plot(x_verdier, y_verdier)
````

En annen kommando som kan brukes er `linspace`. Under lages 201 verdier, der 0 er første verdi, og 5 er siste verdi, og så er det like store sprang mellom verdiene

````python
x_tall = np.linspace(0,5,201)
y_tall = x_tall**2
plt.plot(x_tall, y_tall)
````
